Importing Dependencies

```python
#importing the libraries
import pandas as pd     #for reading data
import numpy as np      
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
#importing the dataset
url = "http://bit.ly/w-data"
study_data = pd.read_csv(url)
```

```python
#first 5 rows of the dataset
study_data.head()
```

Visuals of the dataset in a 2D graph to understand the trend

```python
#showing the distribution of scores in graph
study_data.plot(x='Hours', y='Scores', style = 'o')
plt.title('Study time vs Score percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage of score')
plt.show()
```

Preparing the data for training

```python
#dividing the data as input and output
X = study_data.iloc[:, :-1].values  
y = study_data.iloc[:,1].values  
```

```python
#check the shape of X
X.shape
```

```python
#check the shape of y
y.shape
```

```python
#spliting the dataset for training and testion purpose
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
#shape of X_train
X_train.shape
```

```python
#shape of X_test
X_test.shape
```

```python
#creating a training model with linear regressor
from sklearn.linear_model import LinearRegression  
lr_model = LinearRegression()  

#for fitting the data on the model
lr_model.fit(X_train, y_train)
print("Training is completed.")
```

```python
# Regression line plotting
line = lr_model.coef_*X + lr_model.intercept_

# Plotting the test data
plt.scatter(X, y)
plt.plot(X, line, color='red');
plt.show()
```

Predicting from the model

```python
y_pred = lr_model.predict(X_test) # Predicting the scores
```

```python
actualVpred_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
actualVpred_df
```

Finding the error according to the prediction

```python
from sklearn import metrics
mse = metrics.mean_squared_error(y_test, y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("Mean squred error: ", mse)
print("Mean absolute error: ", mae)
print("Root mean squred error: ", rmse)
```

Prediction for 9.25 hours of study

```python
# prediction for given value
predicted_score = lr_model.predict([[9.25]])
print(predicted_score)
```

The predicted score is 93.6917 for study hour of 9.25
